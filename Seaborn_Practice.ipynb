```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
dataset1 = sb.load_dataset("flights")
sb.relplot(x="passengers", y="month", data= dataset1)
```

```python
ds1 = sb.load_dataset("flights")
sb.relplot(x="passengers", y="month",hue="year",data= ds1)
```

```python
ds2 = sb.load_dataset("tips")
sb.relplot(x="time", y="tip", data=ds2, kind="line", hue="day")
```

```python
sb.catplot(x="day", y="total_bill",data=ds2)
```

```python
sb.catplot(x="day", y="total_bill",data=ds2, kind="violin")
```

```python
sb.catplot(x="day", y="total_bill",data=ds2, kind="boxen")
```
